```python
jobs = ['u-aq226', 'u-ar162']
dir   = ['data/ocean/' + i + '/' for i in jobs]
lat = 26
```

```python
def load_varname(dir, varname = 'Meridional Stream-Function: Atlantic'):
    files = listdir_path(dir)
    
    def openFile(file):
        cube = iris.load(file, varname)
        if type(cube) == iris.cube.CubeList: cube = cube[0] 
        return(cube)
    
    cube = [openFile(f) for f in files]
    cube = iris.cube.CubeList(cube).concatenate()
    return cube
```

```python
AMOC = [load_varname(d) for d in dir]
```

```python
def find_amoc_at_x(amoc):
    amoc = amoc.collapsed('depth', iris.analysis.MAX)
    index = np.absolute(amoc.coord('latitude').points - lat)
    index  = np.where(index == index.min())[0][0]

    return amoc[:, index].data[0][0]

def find_AMOCS_at_x(AMOC):
    return np.array([find_amoc_at_x(amoc) for amoc in AMOC])

AMOC_TS = [find_AMOCS_at_x(amoc) for amoc in AMOC]
```

```python
## plot setup
both = [i.shape[0] for i in AMOC_TS]
Len = [np.min(both), np.max(both)]

def annotate(loc):
    plt.gca().set_ylabel('Sv', fontsize=16)
    plt.legend(loc = loc, bbox_to_anchor = (0.5, -0.05),
               fancybox = True, shadow = True, ncol = 2)

## plot jobs
for amoc, title in zip(AMOC_TS, jobs):
        plt.plot(np.arange(1979, 1979 + len(amoc)), amoc, label = title)


annotate('upper right')
    
## plot difference    
dAMOC = AMOC_TS[1][:Len[0]] - AMOC_TS[0][:Len[0]]
    
ax2 = plt.gca().twinx()
plt.plot(np.arange(1979, 1979 + Len[1]), np.repeat(0.0, Len[1]), 'k--')

plt.plot(np.arange(1979, 1979 + len(dAMOC)), dAMOC, 'r', label = 'difference')
annotate('upper left')

## finish plot
plt.grid(True)    
plt.axis('tight')
```
